```python
import json, os, requests
from requests import request
import pandas as pd
import numpy as np
import time
import pandas_read_xml as pdx
import xmltodict
```

```python
# all record request

userCik = '0001849164'
url = f'https://data.sec.gov/submissions/CIK{userCik}.json'

req = request(url=url, method='GET', headers={"User-Agent": "Mozilla/5.0"})
toplvlData = req.json()['filings']['recent']
toplvlData
```

```python
# toplvl
topLvlDf = pd.DataFrame.from_dict(toplvlData)
topLvlDf = topLvlDf.reset_index()

usableNum = []
usablePrimary = []
for i in topLvlDf.iterrows():
    formData = i[1]['form']
    if  formData == '4':
        usableNum.append(i[1]['accessionNumber'])
        usablePrimary.append(i[1]['primaryDocument'])

print(usablePrimary)
```

```python
for i in usableNum:
    time.sleep(0.1)
    #settup for req
    part1 = i.replace('-', '')
    part2 = usablePrimary[usableNum.index(i)].split('/')[1]
    combinedParts = f'https://www.sec.gov/Archives/edgar/data/{userCik}/{part1}/{part2}'

    #req sending
    iterativeReq= request(url=combinedParts, method='GET', headers={"User-Agent": "Mozilla/5.0"})
    print(iterativeReq.text)
```
